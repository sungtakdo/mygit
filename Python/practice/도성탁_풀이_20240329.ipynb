```python
# 1. 고예찬
# 문제 : 축하합니다. 당신은 공공기관에 출근하게 되었습니다. 당신의 직장은 차량 5부제를 시행하고 있습니다.
#     만약 차량 끝번호가  1,6이라면 월요일에
#                         2,7이라면 화요일에
#                         3,8이라면 수요일에
#                         4,9라면 목요일
#                         5,0이면 금요일에 쉽니다.
#     당신의 차량 끝번호와 오늘이 무슨 요일인지 입력받아 당신의 차량이 쉬는 날인지 아니면 운전해도 되는 날인지 알려주는 코드를 만들고자 합니다.
#     만들어주세요 화이팅.
```

<img src = "https://drive.google.com/uc?id=16ipxCjYbC4ztlgNtw0XeRhmiOiYhdB87">

```python
def car(number, day):
    if number in [1,6] and day in ["월요일", "월"]:
        return True
    elif number in [2,7] and day in ["화요일", "화"]:
        return True
    elif number in [3,8] and day in ["수요일", "수"]:
        return True
    elif number in [4,9] and day in ["목요일", "목"]:
        return True
    elif number in [5,0] and day in ["금요일", "금"]:
        return True
    else :
        return False

number = int(input("차량 끝 번호 입력 : "))
day = input("요일 입력 : ")

if car(number, day):
    print("오늘은 쉬는 날")
else :
    print("차량 운전 가능")
```

```python
# 9. 안연수
# 문제 : 중고차 구매하기
#         차량 종류는 bmw , benz 중 하나, 배기량은 4395cc 이상, 가격은 9천만원 이하, 운행거리가 20km이하면 구매
#         아니면 은행 적금하기
```

<img src = "https://drive.google.com/uc?id=1eaIZSww2LrioaCDbh2fRYSpWE0vUDPPK">

```python
def car(name, power, price, distance):
    if(name in ["BMW","bmw","비엠더블유","BENZ","benz", "벤츠"]) and (power >= 4395) and (price <= 9000) and (distance <= 20):
        return True
    else:
        return False

name = input("차종 입력 : ")
power = float(input("배기량 입력(cc) :  "))
price = float(input("가격 입력(만원) : "))
distance = float(input("운행거리 입력(km) : "))

if car(name, power, price, distance) == True:
    print("구매")
else:
    print("저금")
```

```python
# 12. 윤홍림
# 문제 : 2개의 숫자를 입력받고, 입력받은 숫자들의 최대공약수와 최소공배수 구하기
```

<img src = "https://drive.google.com/uc?id=1oAou7dakRaZhChSFIsY2Hq2ALQxMlqTG">

```python
import math
a = int(input("숫자 입력 : "))
b = int(input("숫자 입력 : "))
if a % b == 0:
    GCD = b
    LCM = a
else:
    GCD = math.gcd(a,b)
    LCM = a*b//GCD

print("최대공약수 : ", GCD)
print("최소공배수 : ", LCM)
```

```python
# 13. 이정현
# 문제 : 서울시민들을 대상으로 출퇴근 거리관련 인터뷰를 진행하려고 한다.
#         인터뷰 대상자 수를 정하고,
#         출퇴근 거리가 40km이상인 경우 나이, 성별, 이름, 이유를 입력받고 거리, 나이, 성별, 이름, 이유를 출력
#         출퇴근 거리가 40km미만인 경우 나이, 성별, 이름 을 입력받고 거리, 나이, 성별, 이름을 출력
```
