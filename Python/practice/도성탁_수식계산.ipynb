```python
priority = {'(': 1, ')': 1, '+': 2, '-': 2, '*': 3, '/': 3}
def in_to_post(infix):
    s = []
    postfix = []
    for i in infix:
        if i == '(':
            s.push(i)
        elif i == ')':
            while s.top() != '(':
                postfix.append(s.pop())
            s.pop()
        elif i in priority:
            while not s.isEmpty():
                if priority[s.top()] >= priority[i]:
                    postfix.append(s.pop())
                else:
                    break
            s.push(i)
        else:
            postfix.append(i)
    while not s.isEmpty():
        postfix.append(s.pop())

    return postfix
```

```python
def post_cal(postfix):
    s = []

    for i in postfix:
        if i in priority:
            num1 = s.pop()
            num2 = s.pop()
            if i == '+':
                s.push(num2 + num1)
            elif i == '-':
                s.push(num2 - num1)
            elif i == '/':
                s.push(num2 / num1)
            elif i == '*':
                s.push(num2 * num1)
        else:
            s.push(float(i))

    return s.pop()
```
