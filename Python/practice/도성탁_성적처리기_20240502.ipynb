```python
import random
```

```python
score = []
for i in range(20):
    a = random.randint(0,100)
    score.append(a)

print(score)
```

```python
print(sum(score)/20)
```

```python
for i  in range(len(score)):
    if score[i] >= 90 and score[i] <= 100 :
        print('A 등급')
    elif score[i] >= 80 and score[i] < 90 :
        print('B 등급')
    elif score[i] >= 70 and score[i] < 80 :
        print('C 등급')
    elif score[i] >= 60 and score[i] < 70 :
        print('D 등급')
    else:
        print('E 등급')
```

```python
class 성적처리기:
    def __init__(self, score):
        self.score = score

    def 평균점수(self):
        print(f"평균 점수는 {sum(score)/20}점")

    def 등급(self):
        for i  in range(len(score)):
            if score[i] >= 90 and score[i] <= 100 :
                print(f'{i+1}번 학생은 A 등급')
            elif score[i] >= 80 and score[i] < 90 :
                print(f'{i+1}번 학생은 B 등급')
            elif score[i] >= 70 and score[i] < 80 :
                print(f'{i+1}번 학생은 C 등급')
            elif score[i] >= 60 and score[i] < 70 :
                print(f'{i+1}번 학생은 D 등급')
            else:
                print(f'{i+1}번 학생은 E 등급')
```

```python
cal = 성적처리기(score)
cal.평균점수()
cal.등급()
```
