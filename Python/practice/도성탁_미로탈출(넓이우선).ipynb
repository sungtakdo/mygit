```python
def bfs(array, start):
    visited = []
    queue = []
    queue.append(start)
    dh = [0, 1, -1, 0]
    dw = [1, 0, 0, -1]
    while queue:
        node = queue.pop(0)
        visited.append((node[0],node[1]))
        for i in range(4):
            y_check = node[1] + dh[i]
            x_check = node[0] + dw[i]
            if 0<= y_check < len(array) and 0<= x_check < len(array):
                if array[y_check][x_check] == '0':
                    queue.append((x_check, y_check))
                elif array[y_check][x_check] == 'x':
                    visited.append((x_check,y_check))
                    return print(visited)
        array[node[1]][node[0]] = '2'
    return print(visited)
```
