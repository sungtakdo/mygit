List 메서드(method)

| |메서드|기능|
|:--|:---:|:--|
|1|append()|리스트 끝에 요소 한개를 추가하기|
|2|clear()|리스트내의 모든 항목 제거|
|3|extend()|리스트 끝에 요소 여러개 추가하기|
|4|insert(인덱스, 값)|인덱스 위치에 값 삽입|
|5|remove()|값의 첫 인스턴스 제거|
|6|count()|인스턴스 개수 반환|
|7|index()|값의 인덱스 반환|
|8|pop(인덱스)|인덱스의 값 반환 및 제거|
|9|reverse()|현재 순서를 뒤집는다(값을 정렬하지는 않음)|
|10|sort()|리스트 내 요소를 순서대로 정렬하기|
|11|copy()|리스트 전체를 복사하여 반환하기|


```python
# append()
alphabet = ['a','b','c','d','e']
alphabet.append('f')
print(alphabet)
```

```python
# clear()
alphabet = ['a','b','c','d','e']
alphabet.clear()
print(alphabet)
```

```python
# extend()
alphabet = ['a','b','c','d','e']
add_list = ['f','g']
alphabet.extend(add_list)
print(alphabet)
```

```python
# insert()
alphabet = ['a','b','c','d','e']
alphabet.insert(1, 'f')
print(alphabet)
```

```python
# remove()
alphabet = ['a','b','c','d','e','b']
alphabet.remove('b')
print(alphabet)
```

```python
# count()
alphabet = ['a','b','c','d','e','c']
count_c = alphabet.count('c')
print(count_c)
```

```python
# index()
alphabet = ['a','b','c','d','e']
D=alphabet.index('d')
print(D)
```

```python
# pop()
alphabet = ['a','b','c','d','e']
alphabet_pop = alphabet.pop(1)
print(alphabet_pop)
print(alphabet)
```

```python
# reverse()      리스트의 순서를 뒤집지만, 값을 정렬하지는 않는다.
alphabet = ['a','d','e','c','b']
alphabet.reverse()
print(alphabet)
```

```python
# sort()
alphabet = ['e','d','a','c','b']
alphabet.sort()
print(alphabet)
```

```python
# copy()
alphabet = ['a','b','c','d','e']
alphabet_copy = alphabet.copy()
print(alphabet_copy)
print(alphabet)
```
